```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('./penguins.csv')
df.head()
```

```python
#
plt.hist(df['flipper_length_mm'].dropna(), bins=20, color='blue', alpha=0.9)
plt.title('Distribution of Flipper Lengths')
plt.show()
```

```python
##Haremos un conteo de machos y hembras por especie 

pf = pd.read_csv('./penguins.csv')

##Seleccionamo las columnas y sus valores 

Datos = pd.crosstab(df['species'], df['sex'])
```

```python
#Grafica de barras
Datos.plot(kind='kde', stacked=True, color=['pink', 'lightblue'])
#Titulo y etiquetas
plt.title('Machos y hembras por especie de pingino')
#nombre de los ejes X,Y
plt.xlabel('especie')
plt.ylabel('Cantidad')
#imprimir o mostra grfica 
plt.show()
```
